```python
data1 = data.drop(['Unnamed: 0',"Temp C"], axis =1)
data1
```

```python
data1.rename({'Solar.R': 'Solar', 'Temp':"Temperature"}, axis=1, inplace = True)
data1
```

```python
#Display data1 missing values count in each column using isnull().sum()
data1.isnull().sum()
```

```python
#Visualize data1 missing values
cols = data1.columns
colours = ['black', 'yellow']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colours),cbar = True)
```

```python
#Find the mean and median values of each numeric column
#imputation of missing value with median
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean()
print("Median of Ozone: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
#Replace the ozone missing values with median value
data1['Ozone'] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
median_solar = data1["Solar"].median()
mean_wind = data1["Wind"].mean()
print("Median of Solar: ", median_solar)
print("Mean of Wind: ", mean_wind)
```

```python
data1['Solar'] = data1['Solar'].fillna(median_solar)
data1.isnull().sum()
```

```python
data1['Wind'] = data1['Wind'].fillna(mean_wind)
data1.isnull().sum()
```

```python
print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python

data1["Weather"] = data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
mu = data1["Ozone"].describe()[1]
sigma = data1["Ozone"].describe()[2]

for x in data1["Ozone"]:
    if ((x < (mu - 3*sigma)) or (x > (mu + 3*sigma))):
        print(x)
        
```

Observations:
 - It is observed that only two outliers are identified using std method
 - In box plot method more no of outliers are identified
 - This is because the assumption of normality is not satified in this column

```python
import scipy.stats as stats

#Create Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(data1["Ozone"], dist="norm", plot=plt)
plt.title("Q-Q Plot for Outlier Detection", fontsize=12)
plt.xlabel("Theoretical Quantiles", fontsize=10)
```

Observations from Q-Q plot
 - The data does not follow normal distribution as the data points are devating significantly away from the red line
 - The data shows a right-skewed distribution and possible outliers
